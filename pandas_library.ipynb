```python
import pandas as pd
```

```python
data = { "cars": ["BMW", "Honda", "Suzuki", "Tata"],
        "models": ["2005", "2010", "2015", "2018"],
         "passenger": ["4", "5", "7", "10"]
        }
type(data)
```

```python
# convert dict ti dataframe
df = pd.DataFrame(data)
df
```

```python
# save df to csv
df.to_csv("cars.csv")
```

```python
# pandas series
a = [1, 7, 2]
myvar = pd.Series(a)
print(myvar)
```

```python
# create labes in pandas with index
a = [1, 7, 2]
myvar = pd.Series(a, index = ["x", "y", "z"])
print(myvar)
```

```python
# access only the value of x
print(myvar["x"])
```

```python
# create a dataframe having two colums calories and duration
fatburn = { "calories" : ["100", "200", "300", "400", "500"],
           "duration" : ["30min", "45min", "50min", "60min", "65min"]
}
df = pd.DataFrame(fatburn)
print(df)
```

```python
# locate only the first row in df
print(df.loc[3])
```

```python
print(df.loc[[0, 1]])
```

```python
# indexing the dataframe with days name
dframe = pd.DataFrame(fatburn, index = ["sunday", "monday", "tuesday", "wednesday"])
print(dframe)
```
